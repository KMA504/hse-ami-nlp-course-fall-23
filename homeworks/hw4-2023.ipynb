3. Отчет должен содержать: нумерацию заданий и пунктов, которые вы выполнили, код решения, и понятное пошаговое описание того, что вы сделали. Отчет должен быть написан в академическом стиле, без излишнего использования сленга и с соблюдением норм русского языка.

## Часть 1. [1 балл] Эксплоративный анализ

1. Приведите статистику по объектам (словам/выражениям) корпуса, а также по типам связей между объектами.
2. Сформулируйте сущность связи между ними и объясните ее причину.

## Часть 2. [4 балла]  От корпуса к графу
1. Создайте граф из тезауруса RuWordNet с помощью библиотеки NetworkX.
2. Выберите 3 пары случайных сущностей из графа. Используя алгоритм поиска кратчайшего пути в графе (например, алгоритм Дейкстры), найдите пути внутри каждой пары.
3. Приведите и проанализируйте полученные пути, обратите внимание на тип связи.

## Часть 3. [4 балла]  Модель определения типа связи
1. На основе вашего графа обучите Node2Vec (https://github.com/eliorc/node2vec) и получите векторное представление каждого из узлов графа.
2. Для каждой связи графа проконкатенируйте эмбеддинги связанных узлов и присвойте ему метку класса, соответствующую типу связи. Разделите выборку на train/val/test в пропорции 80%:10%:10% с сохранением распределения по меткам класса.
3. Обучите модель логистической регрессии для определения типа связи и оцените ее качество.
